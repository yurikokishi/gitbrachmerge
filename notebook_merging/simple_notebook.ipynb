```python
import numpy as np
import altair as alt
import pandas as pd
```

```python
x = np.linspace(0,6)
y = np.sin(x)
df = pd.DataFrame({'x': x, 'y': y})
```

```python
alt.Chart(df
        ).mark_point(
            color='red'
        ).encode(
            x='x:Q',
            y='y:Q')
```
